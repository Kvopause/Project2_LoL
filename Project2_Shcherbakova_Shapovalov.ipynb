```python
all_teams = {}
#заполнение словаря со статистикой w/l для каждой команды
for key in dict_lol.keys():
    #проверка наличия ключа (названия команды) в словаре
    if dict_lol[key]['team1_total']['teamName'] not in all_teams.keys():
        all_teams[(dict_lol[key]['team1_total']['teamName'])] = {'wins': '0', 'loses': '0'}
    if dict_lol[key]['team2_total']['teamName'] not in all_teams.keys():
        all_teams[(dict_lol[key]['team2_total']['teamName'])] = {'wins': '0', 'loses': '0'}
    #заполнение значений для каждой команды
    if (dict_lol[key]['team1_total']['result'] == '1'):
        all_teams[dict_lol[key]['team1_total']['teamName']]['wins'] = int(all_teams[dict_lol[key]['team1_total']['teamName']]['wins']) + 1
    else: 
        all_teams[dict_lol[key]['team1_total']['teamName']]['loses'] = int(all_teams[dict_lol[key]['team1_total']['teamName']]['loses']) + 1
    if (dict_lol[key]['team2_total']['result'] == '1'):
        all_teams[dict_lol[key]['team2_total']['teamName']]['wins'] = int(all_teams[dict_lol[key]['team2_total']['teamName']]['wins']) + 1
    else: 
        all_teams[dict_lol[key]['team2_total']['teamName']]['loses'] = int(all_teams[dict_lol[key]['team2_total']['teamName']]['loses']) + 1
print (all_teams)
```

```python
#Нахождение винрейта для каждой команды
for key in all_teams.keys():
    all_teams[key]['winrate'] = round(int(all_teams[key]['wins']) / (int(all_teams[key]['wins']) + int(all_teams[key]['loses'])) * 100, 1)
print (all_teams)
```

```python
#Создание двух списков с названиями команд и винрейтом
teams_names = []
teams_wr = []
teams_games = []
for key in all_teams.keys():
    teams_names.append(key)
    teams_wr.append(float(all_teams[key]['winrate']))
    teams_games.append(int(all_teams[key]['wins']) + int(all_teams[key]['loses']))
#print (teams_names)
#print (teams_wr)
#print (teams_games)
#Сортировка
zipped_names_wr_games = sorted(zip(teams_names, teams_wr, teams_games))
teams_names, teams_wr, teams_games = zip(*zipped_names_wr_games)
#print (teams_names)
#print (teams_wr)
#print (teams_games)
```
